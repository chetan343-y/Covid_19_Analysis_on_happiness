```python
# import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import the dataset
dataset=pd.read_csv('covid19_Confirmed_dataset.csv')
dataset.head()
```

```python
dataset.shape
```

```python
# delete the useless columns
df=dataset.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
# aggregate the rows by the country
corona_dataset_aggregated =dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape

```

```python
print(corona_dataset_aggregated.loc["China"].dtypes)
```

```python
print(corona_dataset_aggregated.loc["China"])
```

```python
corona_dataset_aggregated = corona_dataset_aggregated.apply(pd.to_numeric, errors='coerce')

corona_dataset_aggregated.loc["China"].T.plot()
```

```python
# visualize data realted to a country
corona_dataset_aggregated.loc["China"].T.plot()
corona_dataset_aggregated.loc["India"].T.plot()
corona_dataset_aggregated.loc["Spain"].T.plot()
plt.legend()
```

```python
# Calculate teh first derivative of the curve
corona_dataset_aggregated.loc['China'].diff().plot()
```

```python
corona_dataset_aggregated.loc['India'].diff().max()
```

```python
corona_dataset_aggregated.loc['Spain'].diff().max()
```

```python
# max infiction of all the countries
countries=list(corona_dataset_aggregated.index)
max_infection_rates=[]

for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated['Max_infection_rates']=max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
# create a new data Frame
corona_data=pd.DataFrame(corona_dataset_aggregated['Max_infection_rates'])
```

```python
corona_data
```
